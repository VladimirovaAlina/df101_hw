```python
import string
import random
```

```python
name = []
score = []
color = ['blue', 'green', 'brown', 'grey']
eyes = []
for i in range(101):
    name.append(''.join(random.SystemRandom().choice(string.ascii_uppercase) for _ in range(5)))
    score.append(random.choice(range(5,50,5)))
    eyes.append(random.choice(color))
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(name, columns=['name'])
```

```python
df['score'] = score
df['eyes'] = eyes
```

```python
df.head()
```

```python
df.to_csv('Info.csv')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(df['eyes'],df['score']);
```

```python
plt.scatter(df['eyes'],df['score']);
```

```python
plt.figure(figsize=(10,7))
plt.plot(df[df['eyes']=='blue']['score'], color='pink',label='for blue')
plt.plot(df[df['eyes']=='green']['score'],label='for green')
plt.plot(df[df['eyes']=='brown']['score'],label='for brown')
plt.plot(df[df['eyes']=='grey']['score'], label='for grey')

plt.xlabel('number of samples')
plt.ylabel('score')
plt.legend(loc='upper right');
```
